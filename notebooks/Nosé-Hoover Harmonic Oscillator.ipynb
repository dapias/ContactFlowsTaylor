# Nosé-Hoover harmonic oscillator. Gibbs distribution using the parameters given by Tuckerman book.

```julia
using PyPlot
```

```julia
include("../Nose-HooverTaylor.jl");
```

```julia
condinic = [0., 1.,0.,1.] #q, p, eta, p_eta;
```

```julia
nsteps = 20000; 
ndescarte = 100; #Descarto los primeros pasos que pueden ser inútiles para la estadística;
```

```julia
vec =NoseHooverIntegration(campoNoseHoover, nsteps, condinic);
```

**vec**  es una tupla de arreglos con entradas `(tiempo,q, p , eta,p_eta)` .
Cada elemento de la tupla es un arreglo de longitud *nsteps*, conteniendo los resultados de la integración de l campo a cada paso de tiempo.


```julia
t = vec[1][ndescarte:end]; q = vec[2][ndescarte:end]; p = vec[3][ndescarte:end]; eta= vec[4][ndescarte:end]; p_eta=vec[5][ndescarte:end];
```

## Espacio fase 

```julia
plt[:figure](1)
plt[:plot](q, p, "g.")
plt[:xlabel]("q")
plt[:ylabel]("p")
```

## p(t) y q(t)

```julia
plt[:figure](1)
plt[:subplot](211)
plt[:plot](t, p, "r.")
plt[:xlabel]("t")
plt[:ylabel]("p")

plt[:subplot](212)
plt[:plot](t, q, "b.")
plt[:xlabel]("t")
plt[:ylabel]("q")

```
